```python
for site in sites:
    site_content = requests.get(site).content
    soup = BeautifulSoup(site_content)
    fullImage = soup.find("div",{"class":"fullImageLink"})
    image_site = "http://www.ssbwiki.com" + fullImage.find("a")["href"]
    r = requests.get(image_site)
    with open("mario/gifs/"+site.split(":")[-1], "wb") as f:
        f.write(r.content)
```
